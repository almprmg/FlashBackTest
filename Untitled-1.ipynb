```python
import pandas as pd
from BackTest_bisc import BackTest
from Stratigy import Stratigy

timeframe = "1h"
timeflow = "5m"
symbol = "AGLDUSDT"
data_low = pd.read_csv(f'C:/Users\dell/Desktop/module Python for Analysis Tradr/data/{timeflow}/{symbol}.csv',index_col=0)
df1= pd.read_csv(r'C:\Users\dell\Desktop\module Python for Analysis Tradr\Backtest_orders\data.csv',index_col=0)
data_low.index = pd.to_datetime(data_low.index)

df1["Date"]  = pd.to_datetime(df1["Date"] ,unit='ms')
df1 = df1.set_index('Date')
class myClass(Stratigy):
    def init(self) -> None:
       super().init()
       self.highest_bot = df1.highest_bot
       self.highest_top = df1.highest_top

    def next(self) -> None:
        super().next()
        data = self.Data.loc[self.Data.Signal == 1]
        signal =  data.index[0] if not data.empty else None

        if signal != None and self.Data.Signal[signal] == 1  :
            self.limet = self.highest_top[signal]
            highest_bot =self.highest_bot[signal]
            self.sl = highest_bot- ((self.limet - highest_bot) *3) 
            self.tp =  highest_bot + ((self.limet - highest_bot) *5) 
            
            self.buy(limet=self.limet ,tp=self.tp,sl= self.sl )
        elif signal != None and  self.Data.Signal[signal]== 2:
            self.Data = self.Data.loc[self.Data.index > signal[0] ]
```

```python
bt =BackTest(df1,data_low,myClass)      
bt.run()

```

```python
import pandas as pd 
from binance.client import Client
import datetime


client = Client('', '')


tickers = client.get_all_tickers()
tickers = [ticker['symbol'] for ticker in tickers ]

end_date = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')[:-3]  # التاريخ الحالي

columns = ['Date','Open','High','Low','Close','Volume','IGNORE','Quote_Volume','Trades_Count','BUY_VOL','BUY_VOL_VAL','x']


def get_klines( symbol,interval, start_date,end_date):
    data = client.get_historical_klines(symbol, interval, start_date, end_date)
    df = pd.DataFrame(data)
    if not df.empty:
        #for column in columns:
        #    df[column] = pd.to_numeric(df[column])


        df.columns = columns
        df["Volume"] = pd.to_numeric(df["Quote_Volume"])
        df["Open"] = pd.to_numeric(df["Open"])
        df["High"] = pd.to_numeric(df["High"])
        df["Low"] = pd.to_numeric(df["Low"])
        df["Close"] = pd.to_numeric(df["Close"])
        for x in range(6,len(columns)):
            del df[columns[x]]
        
        df["Date"] = pd.to_datetime(df["Date"], unit= "ms")
        df = df.set_index("Date")
    return df


```

```python
import os
from datetime import datetime, timedelta

# المسار الذي ترغب في التحقق منه
directory = f"C://Users//dell//Desktop//module Python for Analysis Tradr//data//5m"

# تاريخ اليوم
today = datetime.now().timetuple()

# قائمة لتخزين أسماء الملفات التي لم يتم تعديلها اليوم
untouched_files = []

# الحصول على قائمة الملفات في المسار المحدد
files = os.listdir(directory)

for file in files:
    # مسار الملف الكامل
    file_path = os.path.join(directory, file)
    
    # تاريخ آخر تعديل للملف
    last_modified = datetime.fromtimestamp(os.path.getmtime(file_path)).timetuple()
    
    # التحقق إذا كان تاريخ آخر تعديل هو اليوم
    if last_modified.tm_hour < today.tm_hour:
        untouched_files.append(file)
```
